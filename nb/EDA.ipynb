```python
def find_max(arr,number_arr):
    ind = []
    val = max(arr)
    print("Highest frequency is  ", val)
    for i in range(len(arr)):
        if arr[i]==val:
            ind.append(i)
    res = []
    for n in ind:
        res.append(number_arr[n])
    print("The number with highest frequency is ", res)
    return res
```

```python
#what are the spread features for each draw?
# max value difference, variance
def find_spread(df,col):
    draw_max = []
    draw_min = []
    for n in range(len(df)):
        draw_max.append(df[col].loc[n,:].max())
        draw_min.append(df[col].loc[n,:].min())
    df = df.join(pd.DataFrame({'draw_max':draw_max}))
    df = df.join(pd.DataFrame({'draw_min':draw_min}))
    df['diff'] = df['draw_max'] - df['draw_min']
    return df
```

```python
def visualize_elbow(X,ks):
    fig,ax = plt.subplots()
    inertias = []
    for k in ks:
        km = KMeans(n_clusters=k, random_state=10).fit(X)
        inertias.append(km.inertia_)
    plt.plot(ks,inertias)
    plt.xticks(ks)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Number of Inertia')
    plt.title('Elbow plot')
X = df.drop('DrawDate', axis=1)
ks = list(range(1,11))
visualize_elbow(X,ks)
```

```python
# generate noise data to train the model again
# instead of generating normal distributed data, I would like to generate random data that assume all numbers have the same probability to be drawn
# for normal distribtuion, start = -70, end=70
import scipy.stats as ss

def generate_noise(start,end,std,size):

    x = np.arange(start,end)
    xU, xL = x + 0.5, x - 0.5
    prob = ss.norm.cdf(xU, scale = std) - ss.norm.cdf(xL, scale = std)
    prob = prob / prob.sum() # normalize the probabilities so their sum is 1
    nums = np.random.choice(x, size = size, p = prob)
    #plt.hist(nums, bins = len(x)) # plot the nums to check the distribtution, here it is not a normal distribution
    return nums

lot = []
power = []
for i in range(100):
    random_lot= generate_noise(1,70,std=100,size=5)
    random_power = generate_noise(1,26,std=50,size=1)
    lot.append(sorted(random_lot))
    power.append(random_power)
df_lot = pd.DataFrame(lot)
df_power = pd.DataFrame(power)
noise = pd.concat([df_lot,df_power],axis=1)
noise.columns = df.columns.to_list()[1:7]
```
